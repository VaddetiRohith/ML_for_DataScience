### Preparing the model

```python
import statsmodels.formula.api as smf
```

```python
df.columns
```

```python
model = smf.ols('MPG ~ HP+ VOL+ SP+ WT', data=df).fit()
model
```

```python
#VOL
vol_model= smf.ols('MPG~VOL', data=df).fit()
vol_model.summary()
```

```python
#WT
wt_model= smf.ols('MPG~WT', data=df).fit()
wt_model.summary()
```

```python
#MPG with VOL & WT
wt_vol_model= smf.ols('MPG~ VOL+ WT', data=df).fit()
wt_vol_model.summary()
```

### calculate VIF

```python
df.columns
```

```python
r_hp = smf.ols('HP~ VOL+ SP+ WT', data=df).fit().rsquared
vif_hp = 1/(1-r_hp)
```

```python
r_vol = smf.ols('VOL~ HP+ SP+ WT', data=df).fit().rsquared
vif_vol = 1/(1-r_vol)
```

```python
r_sp = smf.ols('SP~ VOL+ HP+ WT', data=df).fit().rsquared
vif_sp = 1/(1-r_sp)
```

```python
r_wt = smf.ols('WT~ VOL+ SP+ HP', data=df).fit().rsquared
vif_wt = 1/(1-r_wt)
```

```python
vif = pd.DataFrame({
    'Feature_names' : ['HP', 'VOL', 'SP', 'WT'],
    'VIF' : [vif_hp, vif_vol,vif_sp, vif_wt]
})
vif
```
